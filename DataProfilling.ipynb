```python
import cProfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python

class DataAnalyzer:
    def __init__(self, numsamples):
        self.numsamples = numsamples
        self.data = None
        self.data_frame = None

    def generate_data(self):
        self.data = np.random.randint(1, 100, size=self.numsamples)
        self.data_frame = pd.DataFrame(self.data,columns=['RandomData'])
        print("Generated Data:")
        print(self.data_frame)

    def analyze_data(self):
        mean_val = self.data_frame['RandomData'].mean()
        median_val = self.data_frame['RandomData'].median()
        std_dev = self.data_frame['RandomData'].std()

        print("\nData Analysis:")
        print(f"Mean: {mean_val}")
        print(f"Median: {median_val}")
        print(f"Standard Deviation: {std_dev}")

    def visualize_data(self):
        plt.hist(self.data_frame['RandomData'], bins=10, edgecolor='black')
        plt.title("Random Data Distribution")
        plt.xlabel("Value")
        plt.ylabel("Frequency")
        plt.show()
```

```python
def main():
    analyzer = DataAnalyzer(numsamples=1000)
    start_time = time.time()
    analyzer.generate_data()
    analyzer.analyze_data()
    analyzer.visualize_data()
    end_time = time.time()
    execution_time = end_time - start_time
    print(f"\nTotal Execution Time: {execution_time} seconds")

if __name__ == "__main__":
    main()
```

```python
cProfile.run('main()')
```
